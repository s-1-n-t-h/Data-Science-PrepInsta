### Importing Libraries

```python
import requests
```

### API URL

```python
username = 's-1-n-t-h'
api_url = 'https://api.github.com/users/'
url = api_url+username
```

### Calling API

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
print(response.url)
```

### Extracting data as Python Object

```python
data = response.json()
```

```python
data
```

### Printing Profile Picture of GitHub Account

```python
from skimage import io
```

```python
img_url = data.get('avatar_url')
```

```python
img = io.imread(img_url)
```

```python
io.imshow(img)
```
