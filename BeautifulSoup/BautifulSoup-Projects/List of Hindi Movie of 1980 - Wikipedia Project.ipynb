## Creating a Beautiful Soup Object from html string

```python
html = response.text
```

```python
data = BeautifulSoup(html,'html.parser')
```

```python
print(data.prettify())
```

## Finding the table of Films

## Finding for the Films Table using class name

```python
table_data = data.find('table',class_='wikitable') # class name found by inspecting the webpage
```

```python
type(table_data)
```

## Finding Rows in the table

```python
rows = table_data.find_all('tr')
```

```python
type(rows)
```

```python
rows
```

## Making a Python List of Films

```python
movies = []

for row in rows:
    currMovie = []
    for cell in row.find_all('td'):
        if cell.string !='\n' and cell.string != None:
            currMovie.append(cell.string)
        else:
            currMovie.append("NA")
    movies.append(currMovie)
```

## Printing the List of Movies

```python
movies
```

```python
warnings.filterwarnings('ignore')
```

## Creating DataFrame for the data scraped

```python
df = pd.DataFrame()

for movie in movies:
    if len(movie) != 0:
        df = df.append({'Title':movie[0],
                        'Director':movie[1],
                        'Genre':movie[3]
                      },ignore_index=True)
        
        
```
