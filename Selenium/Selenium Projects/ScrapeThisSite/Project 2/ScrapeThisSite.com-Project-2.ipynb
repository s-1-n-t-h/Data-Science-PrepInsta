## Importing Libraries 

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd 
import numpy as np
```

## Creating Driver Instance

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

## Navigation to Page

```python
driver.get('https://scrapethissite.com/pages/ajax-javascript/')
```

## Creating a selenium object to click on hyperlink

```python
link_btn = driver.find_element(By.LINK_TEXT,"2010")
drive.save_screenshot('before.png')
```

### finding tag by ID 

```python
search_box = driver.find_element(By.ID,'q')
```

### sending data to enter into text box and submit data

```python
search_box.send_keys("New York Rangers")
search_box.submit()
```

```python
table_data = driver.find_element(By.CLASS_NAME,'table')
table_data.tag_name
```

### Creating Soup Object 

```python
page_data = BeautifulSoup(driver.page_source)
team_data = page_data.table
print(type(team_data))
team_data
```

### finding table by class

```python
table = page_data.find(class_='table')
```

```python
type(page_data)
```

```python
type(table)
```

### extracting rows of table 

```python
rows = table.find_all('tr')
```

```python
rows
```

### Creating Python List of data

```python
data = []
for row in rows:
    current_row = []
    for cell in row:
        if cell.string!='\n' and cell.string !=None:
            current_row.append(cell.string.strip('\n').strip(' ').strip('\n'))
        
    data.append(current_row)
            
```

```python
data
```

## Creating Pandas DataFrame  

```python
df = pd.DataFrame(data[1:],columns=data[0] )
```
