```python
driver.get("https://www.scrapethissite.com/pages/forms/")

team_list = ['Buffalo Sabres', 'Calgary Flames', 'New York Rangers', 'Calgary Flames']
result_list = []

for team in team_list:
    text_box = driver.find_element(By.ID, 'q')
    text_box.clear()
    text_box.send_keys(team)
    text_box.submit()
    
    page_data = BeautifulSoup(driver.page_source)
    team_data = page_data.table
    df = pd.read_html(str(team_data), header = 0)
    result_list.append(df[0])

result_list
```

```python
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
driver.get("https://www.flipkart.com/")
```

```python
text_box = driver.find_element(By.CLASS_NAME,'_3704LK')
text_box.send_keys('Mobile Phones')
text_box.submit()
# can use this
# import time
# time.sleep(5)
driver.implicitly_wait(3)
```

```python
thisdict = {}

product_names = driver.find_elements(By.CLASS_NAME, '_4rR01T')
product_prices = driver.find_elements(By.CLASS_NAME, '_1_WHN1')


for index, item in enumerate(product_names, 0):
    thisdict.update({item : product_prices[index]})


for x, y in thisdict.items():
    print(x.text, y.text)
```

```python
product_names = driver.find_elements(By.CLASS_NAME, '_4rR01T')
for product in product_names:
    print(product.text)
```
