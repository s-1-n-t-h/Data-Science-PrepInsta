```python
# Relative xPath
# https://www.scrapethissite.com/pages/forms/?q=New+York+Rangers
# //table
# //input (will give two results)
# //input[@placeholder='Search for Teams']
# //input[@value='New York Rangers']
# //input[@value='Search']
# //input[@class='btn btn-primary']
# //input[@class='btn'] (wont work as two classes)
# //* (gives all result)
# //*[@class="container"] (gives 3 results)
# //nav/div[@class="container"] (gives one result)
# //a (gives 8 results)
# //a[@href='pages'] (wont work)
# //a[@href='/pages/'] (will work)
# //li (gives many results)
# //li[@id='nav-login'] (one result)
# //img[@src='/static'] (Wont find as need whole text)
# 
# //*[contains(@src,'/static')]
# //*[contains(@class,'btn')]
# //*[contains(text(),'L')]
# //*[contains(text(),'Login')]
# //a[contains(text(),'Login')]
# //a[contains(., 'Login')] 
# //*[contains(., 'Take')] (selects all items)
# //*[contains(text(), 'Take')]
# //input[@type='submit' and @value='Search']
# //input[@value='SEARCH' or @value='Search']
# //*[starts-with(@type,'text')]
# //*[contains(text(),'L')]
# //*[contains(text(),'L')]//ancestor::li
# //*[contains(text(),'L')]//ancestor::div (gives more results as ancestors outer classes)
# //a
# //a//child::i
# preceeding, following sibling etc read more here -https://www.guru99.com/xpath-selenium.html
```
