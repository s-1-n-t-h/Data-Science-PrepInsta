## GET Request for query under tv category

```python
import requests

api_key = 'e5134c4a5a7554660a16ab36ea430c03'
api_url = 'https://api.themoviedb.org/3'


params = {'query':'game','api_key':api_key}
headers = {'Accept':'application/json'}

response = requests.get(api_url+'/search/tv',headers = headers,params = params)

# creating python dict object from json
data = response.json() # == data['results']
```

```python
# using dict.get() method to get value from a key
results = data.get('results')
results
```

### Names of the TV Shows

```python
for result in results:
    print(result['name'])
```

```python
print(response.url)
```
